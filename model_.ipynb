```python
def trapezoid(f,t0,t1):
    n = len(f)
    return ((t1-t0)/n) * (f[0]/2 + f[-1]/2 + np.sum(f[1:-1]))
```

```python

N = 200
a, b = [0, 10]
dt, T = 0.02, 30

source_args = {'source_type': 'normal', 'source_mu': 5, 'source_std': 0.05}
theta_args = {'theta_value': 4.5, 'theta_moving': False, 'theta_speed': 0.00, 
              'theta_noise': True, 'theta_noise_type': 'normal', 'theta_noise_std': 0.025}
s_args = {'s_value': 1, 's_noise': False, 's_noise_type': 'normal', 's_noise_std': 0.05}

solver = model(N,a,b,dt,T)
simulations = 15
mu_bar = np.zeros(simulations)
var_bar = np.zeros(simulations)
rho_bar = np.zeros(simulations)
t0 = 1
t1 = 30
for i in range(simulations):
    
    solver.solve("solve_implicit", theta_args, s_args, source_args)
    mu = solver.MU
    var = solver.VAR
    rho = solver.RHO
    t0_i = int(t0/dt)
    t1_i = int(t1/dt)
    mu_bar[i] = 1/(t1-t0) * trapezoid(mu[t0_i:t1_i],t0,t1)
    var_bar[i] = 1/(t1-t0) * trapezoid(var[t0_i:t1_i],t0,t1)
    rho_bar[i] = 1/(t1-t0) * trapezoid(rho[t0_i:t1_i],t0,t1)
plt.figure(figsize=(14,5))
plt.subplot(1,3,1)
plt.title("Mu")
plt.plot(mu_bar, '.-')
plt.subplot(1,3,2)
plt.title("Var")
plt.plot(var_bar, '.-')
plt.subplot(1,3,3)
plt.title("Rho")
plt.plot(rho_bar, '.-')
plt.show()
```

```python
t0 = 1
t1 = 30
t0_i = int(t0/dt)
t1_i = int(t1/dt)

theta_args = {'theta_value': 4.5, 'theta_moving': False, 'theta_speed': 0.00, 
              'theta_noise': True, 'theta_noise_type': 'normal', 'theta_noise_std': 0.025}
sigs = np.arange(0.1,0.9,0.05)
fardeau_evolutive_moyenne = np.zeros(len(sigs))
for i in range(len(sigs)):
    solver.sigma = sigs[i]
    solver.solve("solve_implicit", theta_args, s_args, source_args)
    temp = solver.S*(solver.MU-solver.THETA)**2
    fardeau_evolutive_moyenne[i] = trapezoid(temp[t0_i:t1_i],t0,t1)
plt.plot(sigs,fardeau_evolutive_moyenne)
```
